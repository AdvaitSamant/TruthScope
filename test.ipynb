```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

BASE_URL = "https://pib.gov.in"

url = "https://pib.gov.in/allRel.aspx"

headers = {
    "User-Agent": "Mozilla/5.0"
}

print("Connecting to PIB website...")

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

articles = []

links = soup.find_all("a", href=True)

print("Collecting article links...")

for a in links:

    if "PressReleasePage.aspx?PRID" in a["href"]:

        title = a.text.strip()

        if title:

            link = BASE_URL + "/" + a["href"]

            try:

                page = requests.get(link, headers=headers)

                article_soup = BeautifulSoup(page.text, "html.parser")

                paragraphs = article_soup.find_all("p")

                content = ""

                for p in paragraphs:

                    text = p.get_text().strip()

                    if len(text) > 30:
                        content += text + " "

                articles.append({
                    "Title": title,
                    "Content": content,
                    "Link": link
                })

                print("Scraped:", title[:60])

                time.sleep(1)

            except:
                continue


df = pd.DataFrame(articles)

print("\nTotal articles scraped:", len(df))

df.to_csv("pib_dataset.csv", index=False)

print("\nDataset saved as pib_dataset.csv")
```
